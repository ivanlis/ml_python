```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
# MATLAB-style interface
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# Object-oriented interface
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```
