# Seaborn: plotting with categorical data 

See: https://seaborn.pydata.org/tutorial/categorical.html

```python
import seaborn as sns
import matplotlib.pyplot as ptl

sns.set(style = "ticks", color_codes = True)
```

## Categorical scatterplots

```python
tips = sns.load_dataset("tips")
```

catplot(): with a random *jitter*.

```python
sns.catplot(x = "day", y = "total_bill", data = tips);
```

```python
sns.catplot(x = "day", y = "total_bill", jitter = False, data = tips);
```

### Swarm

```python
sns.catplot(kind = 'swarm', x = "day", y = "total_bill", data = tips);
```

### Hue semantics

```python
sns.catplot(kind = 'swarm', x = "day", y = "total_bill", hue = "sex", data = tips);
```

###  Ordering

```python
sns.catplot(x = 'smoker', y = 'tip', order = ['No', 'Yes'], data = tips);
```

### Vertical categorical axis

```python
sns.catplot(x = 'total_bill', y = 'day', hue = 'time', kind = 'swarm', data = tips);
```

## Distributions of observations within categories

### Boxplots

```python
sns.catplot(kind = 'box', x = 'day', y = 'total_bill', data = tips);
```

```python
sns.catplot(kind = 'box', x = 'day', y = 'total_bill', hue = 'smoker', data = tips);
```

### Boxenplot

```python
diamonds = sns.load_dataset('diamonds')
```

```python
sns.catplot(kind = 'boxen', x = 'color', y = 'price', data = diamonds.sort_values('color'));
```

### Violinplot

```python
sns.catplot(x = 'total_bill', y = 'day', hue = 'time', kind = 'violin', data = tips);
```

```python
sns.catplot(data = tips, kind = 'violin', x = 'day', y = 'total_bill', hue = 'sex', split = True);
```

## Statistical estimation within categories

### Bar plots

```python
titanic = sns.load_dataset("titanic")
```

```python
sns.catplot(data = titanic, kind = 'bar', x = 'sex', y = 'survived', hue = 'class');
```

```python
titanic.head()
```

```python
sns.catplot(x = 'deck', kind = 'count', palette = "ch:.25", data = titanic);
```

### Point plot

```python
sns.catplot(data = titanic, kind = 'point', x = 'sex', y = 'survived', hue = 'class');
```

## Wide-form data

```python
iris = sns.load_dataset('iris')

iris.head()
```

```python
sns.catplot(data = iris, orient = 'h', kind = 'box');
```

## Facets

```python
sns.catplot(x = 'day', y = 'total_bill', hue = 'smoker',
           col = 'time', aspect = 0.6, kind = 'swarm', data = tips);
```

### Axes-level properties

```python
g = sns.catplot(data = titanic[titanic['fare'] > 0],
               kind = 'box', orient = 'h', height = 1.5, aspect = 4,
               x = 'fare', y = 'survived', row = 'class')
g.set(xscale = 'log');
```
