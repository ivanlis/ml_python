```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = 'darkgrid')

tips = sns.load_dataset('tips');
tips.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'smoker', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips)
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = np.random.randn(500).cumsum()))
df.head()
```

```python
g = sns.relplot(data = df, x = "time", y = "value", kind = "line")
g.fig.autofmt_xdate()
```

```python
# Uncertainty
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', kind = 'line');
```

```python
# Without confidence intervals
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', kind = 'line', ci = None);
```

```python
# Use the standard deviation instead of the confidence interval
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', kind = 'line', ci = 'sd');
```

```python
# Semantic mappings
sns.relplot(data = fmri, kind = 'line', x = 'timepoint', y = 'signal', hue = 'event');
```

```python
sns.relplot(data = fmri, kind = 'line', x = 'timepoint', y = 'signal', style = 'event', hue = 'region');
```

```python
# Separate by SAMPLING UNITS
sns.relplot(data = fmri.query("event == 'stim'"), kind = 'line', 
            x = 'timepoint', y = 'signal', hue = 'region',
            units = 'subject', estimator = None) # divide by sampling units, no aggregation 
```

```python
# Multiple facets
sns.relplot(data = tips,
           x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time')
```

```python
sns.relplot(data = fmri.query("region == 'frontal'"), kind = 'line',
           x = 'timepoint', y = 'signal', hue = 'event', style = 'event',
           col = 'subject', 
           col_wrap = 5, height = 3, aspect = 0.75, linewidth = 2.5)
```
