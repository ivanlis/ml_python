```python
import numpy as np
import matplotlib.pyplot as plt


rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
fig = plt.figure()
ax = plt.axes()

ax.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept = True)
model
```

```python
X = x.reshape((len(x), 1))
X.shape
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
xfit = np.linspace(-1, 11, 50)
Xfit = xfit[:, np.newaxis]
Xfit.shape
```

```python
yfit = model.predict(Xfit)

fig = plt.figure()
ax = plt.axes()
ax.scatter(x, y)
ax.plot(xfit, yfit);
```
