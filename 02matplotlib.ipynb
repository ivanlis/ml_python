```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
# Labeling
fig = plt.figure()
axes = plt.axes()

axes.plot(x, np.sin(x), color='r', linestyle='dashed', label='sin(x)');
axes.plot(x, np.cos(x), color='b', linestyle='dotted', label='cos(x)');
axes.set_ylim(-1.5, 1.5)
axes.legend();
axes.set_title('Sine and Cosine')
axes.set_xlabel('x')
axes.set_ylabel('y')
```

```python
# Scatter Plots
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

fig = plt.figure()
ax = plt.axes()
ax.scatter(x, y, s=sizes, c=colors, alpha=0.3, cmap='viridis');
```

```python
# Plot a dataset
from sklearn.datasets import load_iris
iris = load_iris()
#iris.data
features = iris.data.T


fig = plt.figure()
ax = plt.axes()

ax.scatter(features[0], features[1], alpha = 0.3,
           s = 100 * features[3], c = iris.target, cmap = 'viridis')

ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1]);
```
