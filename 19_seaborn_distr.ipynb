# Visualizing distributions

See: https://seaborn.pydata.org/tutorial/distributions.html

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(color_codes = True)
```

## Univariate distributions

```python
np.random.seed(2008)
x = np.random.normal(size = 500)
```

```python
sns.distplot(x);
```

```python
sns.distplot(x, kde = False, rug = True, bins = 20);
```

```python
sns.distplot(x, hist = False, kde = True, rug = True);
```

```python
sns.kdeplot(x, shade = True);
```

## Fitting parametric distributions

```python
x = np.random.gamma(6, size = 200)
```

```python
sns.distplot(x, kde = False, fit = stats.gamma);
```

## Pairwise relationships in a dataset

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris);
```

### Working with PairGrid directly

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels = 6);
```
